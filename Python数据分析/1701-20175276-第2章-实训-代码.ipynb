# 实训1 创建数组并进行运算

```python
import numpy as np

#第一步
First = np.arange(0,1,0.01)#生成均匀数组
print("数值范围为0-1，间隔为0.01的数组：\n",First)
```

```python
#第二步
Second = np.random.randn(1,100)#生成正态分布随机数
print("100个服从正态分布的随机数为:",Second)
```

```python
#第三步
Second.resize(100)
#四则运算
HeRes = Second+First
print("两数组和的结果",HeRes)
CharRes = Second-First
print("两数组差的结果",CharRes)
JiRes = Second*First
print("两数组积的结果",JiRes)
ShangRes = First/Second
print("两数组商的结果",ShangRes)
```

```python
#第四步
#简单数学统计
print("最大值",np.max(Second))
print("最小值",np.min(Second))
print("所有元素和",np.sum(Second))
print("所有元素积",np.prod(Second))
print("标准差",np.std(Second))
print("方差",np.var(Second))
print("平均值",np.mean(Second))
print("中位数",np.median(Second))


#去重
res = np.unique(Second)
#print("去重后的数组",res)

#重复
res = Second.repeat(2)
#print("重复后的数组",res)
#排序方法
tmp = Second
Second.sort()
#print('sort排序后的数组',Second)

Second = tmp
temp = Second.argsort()
#print('argsort排序后的数组',temp)

res = np.lexsort((First,Second))
res = list(zip(First[res],Second[res]))
print('lexsort排序后的数组',res)

```

# 实训2 国际象棋棋盘

```python
chessboard = np.zeros((8,8))

for i in range(8):
    if(i%2==0):
        chessboard[i] = 1
    else:
        chessboard[:,i] = 1
print("棋盘")
print(chessboard)
```
