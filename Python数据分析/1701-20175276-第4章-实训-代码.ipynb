```python
#Training_Loginfo文件借款成立时间和用户登录时间提取 year,month,week
LoginfoListing = pd.to_datetime(Training_Loginfo.Listinginfo1)
LoginfoLog = pd.to_datetime(Training_Loginfo.LogInfo3)
tmp = {
    '借款成立时间（年）':[i.year for i in LoginfoListing],
    '借款成立时间（月）':[i.month for i in LoginfoListing],
    '借款成立时间（周）':[i.week for i in LoginfoListing],
    '用户登录时间（年）':[i.year for i in LoginfoLog],
    '用户登录时间（月）':[i.month for i in LoginfoLog],
    '用户登录时间（周）':[i.week for i in LoginfoLog]
}
Loginfo_Info = pd.DataFrame(tmp)
print(Loginfo_Info)

#Training_Master文件借款成立时间提取 year,month,week
MasterListing = pd.to_datetime(Training_Master.ListingInfo)

tmp = {
    '借款成立时间（年）':[i.year for i in MasterListing],
    '借款成立时间（月）':[i.month for i in MasterListing],
    '借款成立时间（周）':[i.week for i in MasterListing]
}
Master_Info = pd.DataFrame(tmp)
print(Master_Info)

#Training_Userupdate借款成立时间和用户信息修改时间提取 year,month,week
UserupdateListing = pd.to_datetime(Training_Userupdate.ListingInfo1)
UserupdateUpdate = pd.to_datetime(Training_Userupdate.UserupdateInfo2)
tmp = {
    '借款成立时间（年）':[i.year for i in UserupdateListing],
    '借款成立时间（月）':[i.month for i in UserupdateListing],
    '借款成立时间（周）':[i.week for i in UserupdateListing],
    '用户登录时间（年）':[i.year for i in UserupdateUpdate],
    '用户登录时间（月）':[i.month for i in UserupdateUpdate],
    '用户登录时间（周）':[i.week for i in UserupdateUpdate]
}
Userupdate_Info = pd.DataFrame(tmp)
print(Userupdate_Info)


#借款成交时间与登录时间
Listing_Log= LoginfoListing-LoginfoLog
#借款成交时间与用户信息更新时间
Listing_User = UserupdateListing-UserupdateUpdate
```

# 实训三：使用分组聚合方法进一步分析用户信息更新表和登录信息表

```python
#实训二
import pandas as pd
import numpy as np
Training_Loginfo = pd.read_csv('./four/Training_Loginfo.csv')  # 读取数据
#编码问题
Training_Master = pd.read_csv("./four/Training_Master.csv",encoding='gbk')
Training_Userupdate = pd.read_csv("./four/Training_Userupdate.csv")


#groupby分组
LoginGroup = Training_Loginfo.groupby(by='Idx',sort=0)
UserupdateGroup = Training_Userupdate.groupby(by='Idx',sort=0)

#Training_Loginfo文件借款成立时间和用户登录时间同一用户的最早与最晚时间
Max_Min_Login = LoginGroup[['Listinginfo1','LogInfo3']].agg([np.max,np.min])
print('Training_Loginfo文件借款成立时间和用户登录时间同一用户的最早与最晚时间:\n',Max_Min_Login)
#Training_Userupdate文件借款成立时间和用户登录时间同一用户的最早与最晚时间
Max_Min_Update =  UserupdateGroup[['ListingInfo1','UserupdateInfo2']].agg([np.max,np.min])
print('Training_Userupdate文件借款成立时间和用户登录时间同一用户的最早与最晚时间:\n',Max_Min_Update)
#Training_Loginfo用户登录次数
print("Training_Loginfo用户登录次数:\n",LoginGroup.size())
#Training_Userupdate用户登陆次数
print("Training_Userupdate用户登录次数:\n",UserupdateGroup.size())
```
