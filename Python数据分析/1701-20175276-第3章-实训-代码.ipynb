# 实训1 分析1996-2015年人口数据特征间的关系

```python
#导入mokuai
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MaxNLocator
from matplotlib.font_manager import FontProperties
#在jupyter notebook上显示图形
%matplotlib inline
#加载数据
data = np.load("populations.npz")
```

```python
#标签为年份 1996-2015年总共二十年
yearlabel = data['data'][:,0][0:20][::-1]

#获得各数据
x = range(0,20)
nianmozongrenkou = data['data'][:,1][0:20][::-1]
nanxingrenkou = data['data'][:,2][0:20][::-1]
nvxingrenkou = data['data'][:,3][0:20][::-1]
chengzhenrenkou = data['data'][:,4][0:20][::-1]
xiangcunrenkou = data['data'][:,5][0:20][::-1]
```

```python
# 中文字体设置
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
#分图设置
flg = plt.figure(num=2,figsize=(14,10),dpi=80)
#折线图
ax1 = flg.add_subplot(2,1,1)
#散点图
ax2 = flg.add_subplot(2,1,2)

#折线图添加线段
ax1.plot(yearlabel,nianmozongrenkou,color = "peru",linestyle="--",label=u"年末总人口")
ax1.plot(yearlabel,nanxingrenkou,color = "b",linestyle="--",label=u"男性人口")
ax1.plot(yearlabel,nvxingrenkou,color = "r",linestyle="--",label=u"女性人口")
ax1.plot(yearlabel,chengzhenrenkou,color = "tan",linestyle="--",label=u"城镇人口")
ax1.plot(yearlabel,xiangcunrenkou,color = "plum",linestyle="--",label=u"乡村人口")

#散点图添加线段
ax2.scatter(yearlabel, nianmozongrenkou, s = 25, c = 'peru', alpha = 0.5,label=u"年末总人口")
ax2.scatter(yearlabel, nanxingrenkou, s = 25, c = 'b', alpha = 0.5,label=u"男性人口")
ax2.scatter(yearlabel, nvxingrenkou, s = 25, c = 'r', alpha = 0.5,label=u"女性人口")
ax2.scatter(yearlabel, chengzhenrenkou, s = 25, c = 'tan', alpha = 0.5,label=u"城镇人口")
ax2.scatter(yearlabel, xiangcunrenkou, s = 25, c = 'plum', alpha = 0.5,label=u"乡村人口")

## 添加横轴标签，竖轴标签，标题
plt.xlabel('年份') 
plt.ylabel('万人')
plt.title('1996-2015年年末人口折线图和散点图')

#设置注释位置
ax1.legend( loc="upper right")
ax2.legend( loc="upper right")

#保存图片
plt.savefig("1996-2015年年末人口折线图和散点图.jpg")
#展示
plt.show()


```

年末总人口随着年份的增加而逐年增长
在这二十年中，男性人口与女性人口都有所增加，且男性人口始终高于女性人口
二十年中，城镇人口迅速增加，农村人口迅速下降

# 实训2 分析1996-2015年人口数据特征间的分布和分散情况

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MaxNLocator
from matplotlib.font_manager import FontProperties
#在jupyter notebook上显示图形
%matplotlib inline
#加载数据
data = np.load("populations.npz")
```

```python
#标签为年份 1996-2015年总共二十年
yearlabel = data['data'][:,0][0:20][::-1]

#获得各数据
x = range(0,20)
nianmozongrenkou = data['data'][:,1][0:20][::-1]
nanxingrenkou = data['data'][:,2][0:20][::-1]
nvxingrenkou = data['data'][:,3][0:20][::-1]
chengzhenrenkou = data['data'][:,4][0:20][::-1]
xiangcunrenkou = data['data'][:,5][0:20][::-1]

# 中文字体设置
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False
#设置子图
flg = plt.figure(num=2,figsize=(14,10),dpi=80)

#男女
ax1 = flg.add_subplot(2,1,1)
#城市农村
ax2 = flg.add_subplot(2,1,2)

#设置横纵坐标
plt.xlabel("年份")
plt.ylabel("万人")

#设置标题
plt.title("1996-2015年男女人口个数与城乡人口")

#添加条形图
ax1.bar([i+1996 for i in range(0,20)],nanxingrenkou,width = 0.4,alpha=0.5,label = '男性人口')
ax1.bar([i+1996+0.4 for i in range(0,20)],nvxingrenkou,width = 0.4,color = 'r',alpha=0.5,label = '女性人口')
#设置注释位置
ax1.legend( loc="upper right")

#添加条形图
ax2.bar([i+1996 for i in range(0,20)],chengzhenrenkou,width = 0.4,alpha=0.5,label = '城市人口')
ax2.bar([i+1996+0.4 for i in range(0,20)],xiangcunrenkou,width = 0.4,color = 'r',alpha=0.5,label = '农村人口')
#设置注释位置
ax2.legend( loc="upper right")

#保存图片
plt.savefig("1996-2015年男女人口个数与城乡人口.jpg")
#展示
plt.show()
```

男性人口与女性人口都在增长，且男性人口始终高于女性人口
城镇人口迅速增加，农村人口迅速下降

```python
#设置子图
flg = plt.figure(num=6,figsize=(14,8),dpi=100)

#设置标题
plt.title("1996- 2005-2015 年男女人口比例与城市人口比例")

#设置颜色分类
color = ["royalblue", "salmon"]
#1996- 2005-2015 年男女人口比例
gender_list = ["男性", "女性"]    # 各部分标签
explode = [0.05, 0] #凸出值
ax1 = flg.add_subplot(2,3,1)#子图1位置
size1 = [100*nanxingrenkou[0]/(nanxingrenkou[0]+nvxingrenkou[0]),100*nvxingrenkou[0]/(nanxingrenkou[0]+nvxingrenkou[0])]#1996年男女各部分比例
ax1.pie(size1, explode=explode, colors=color, labels=gender_list, labeldistance=1.2, autopct="%1.2f%%", shadow=False, startangle=90, pctdistance=0.6,)

ax2 = flg.add_subplot(2,3,2)#子图2位置
size2 = [100*nanxingrenkou[9]/(nanxingrenkou[9]+nvxingrenkou[9]),100*nvxingrenkou[9]/(nanxingrenkou[9]+nvxingrenkou[0])]#2005年男女各部分比例
ax2.pie(size2, explode=explode, colors=color, labels=gender_list, labeldistance=1.2, autopct="%1.2f%%", shadow=False, startangle=90, pctdistance=0.6,)

ax3 = flg.add_subplot(2,3,3)#子图3位置
size3 = [100*nanxingrenkou[19]/(nanxingrenkou[19]+nvxingrenkou[19]),100*nvxingrenkou[19]/(nanxingrenkou[19]+nvxingrenkou[19])]#2015年男女各部分比例
ax3.pie(size3, explode=explode, colors=color, labels=gender_list, labeldistance=1.2, autopct="%1.2f%%", shadow=False, startangle=90, pctdistance=0.6,)

#1996- 2005-2015 年城村人口比例
city_list = ["城市","农村"]
ax4 = flg.add_subplot(2,3,4)#子图4位置
size4 = [100*chengzhenrenkou[0]/(chengzhenrenkou[0]+xiangcunrenkou[0]),100*xiangcunrenkou[0]/(chengzhenrenkou[0]+xiangcunrenkou[0])]#1996年城乡各部分比例
ax4.pie(size4, explode=explode, colors=color, labels=city_list, labeldistance=1.2, autopct="%1.2f%%", shadow=False, startangle=90, pctdistance=0.6,)
ax5 = flg.add_subplot(2,3,5)#子图5位置
size5 = [100*chengzhenrenkou[9]/(chengzhenrenkou[9]+xiangcunrenkou[9]),100*xiangcunrenkou[9]/(chengzhenrenkou[9]+xiangcunrenkou[9])]#2005年城乡各部分比例
ax5.pie(size5, explode=explode, colors=color, labels=city_list, labeldistance=1.2, autopct="%1.2f%%", shadow=False, startangle=90, pctdistance=0.6,)
ax6 = flg.add_subplot(2,3,6)#子图6位置
size6 = [100*chengzhenrenkou[19]/(chengzhenrenkou[19]+xiangcunrenkou[19]),100*xiangcunrenkou[19]/(chengzhenrenkou[19]+xiangcunrenkou[19])]#2015年城乡各部分比例
ax6.pie(size6, explode=explode, colors=color, labels=city_list, labeldistance=1.2, autopct="%1.2f%%", shadow=False, startangle=90, pctdistance=0.6,)

#保存图片
plt.savefig("1996- 2005-2015 年男女人口比例与城市人口比例.jpg")
#展示
plt.show()
```

二十年中，男性人口和女性人口的比例几乎没有改变，且基本是男性多于女性2个百分点左右
城乡结构变化剧烈，城镇人口迅速增加，最终超过农村人口13个百分点，达到56%

```python
plt.figure(figsize=(14,8),dpi=100)#设置子图
xlabel = ['男性人口','女性人口','城市人口','乡村人口'] #横坐标
gdp = [nanxingrenkou,nvxingrenkou,chengzhenrenkou,xiangcunrenkou]#数据
plt.boxplot(gdp,notch = True,labels = xlabel,meanline = True)#添加箱形图
plt.title("1996-2015 年人口数据各个特征箱线图")#设置标题
#保存图片
plt.savefig("1996-2015 年人口数据各个特征箱线图.jpg")
#展示
plt.show()
```
